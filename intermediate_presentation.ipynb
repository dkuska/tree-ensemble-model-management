```python
from utils.treelite import (
    compare_models,
    get_model_jsons,
    get_node_list,
    get_counts_of_unique_nodes,
)
from utils.visualizations import (
    plot_distribution_duplicate_ratio,
    plot_bars_unique_nodes,
)
```

```python
model_dir_daily_10_trees = "models/sklearn/daily/10_trees"
model_dir_daily_100_trees = "models/sklearn/daily/100_trees"
model_dir_5days_10_trees = "models/sklearn/5days/10_trees"
model_dir_5days_100_trees = "models/sklearn/5days/100_trees"
```

```python
daily_model_10_trees = get_model_jsons(model_dir_daily_10_trees)
daily_model_100_trees = get_model_jsons(model_dir_daily_100_trees)

fiveday_model_10_trees = get_model_jsons(model_dir_5days_10_trees)
fiveday_model_100_trees = get_model_jsons(model_dir_5days_100_trees)
```

```python
def wrapper(models, exact, sequentially):
    duplicate_data = compare_models(models, sequentially=sequentially, exact=exact)
    duplicate_ratio_dist = plot_distribution_duplicate_ratio(
        duplicate_data, "duplicate_ratio"
    )

    node_list = get_node_list(models, exact=exact)
    unique_node_counts = get_counts_of_unique_nodes(node_list)
    unique_nodes_dist = plot_bars_unique_nodes(node_list, unique_node_counts)

    return duplicate_ratio_dist, unique_nodes_dist
```

## Daily Models

### 10 Trees

```python
wrapper(daily_model_10_trees, exact=False, sequentially=False)
```

### 100 Trees

```python
wrapper(daily_model_100_trees, exact=True, sequentially=True)
```

## 5 Day Models

### 10 Trees

```python
wrapper(fiveday_model_10_trees, exact=True, sequentially=True)
```

### 100 Trees

```python
wrapper(fiveday_model_100_trees, exact=True, sequentially=True)
```
