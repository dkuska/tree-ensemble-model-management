```python
import json
import pickle
import treelite
from treelite import Model
from copy import deepcopy
```

```python
model_path = (
    "models/sklearn/daily/10_trees/..."
)
```

```python
with open(model_path, "rb") as f:
    model = pickle.load(f)

# Load Model into Treelite
treelite_model = treelite.sklearn.import_model(model)
# Load JSON Representation
treelite_model_json = json.loads(treelite_model.dump_as_json(pretty_print=False))
```

```python
def fix_model_json(model_json):
    fixed_json = deepcopy(model_json)

    for tree in fixed_json["trees"]:
        tree["root_id"] = 0

    return fixed_json
```

```python
fixed_json = fix_model_json(treelite_model_json)

reconstructed_model = Model.import_from_json(json.dumps(fixed_json))
reconstructed_model_json = json.loads(
    reconstructed_model.dump_as_json(pretty_print=False)
)
```
